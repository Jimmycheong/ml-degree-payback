# Notebook

```python
import os 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv_filepath = os.getcwd() + "/degrees-that-pay-back.csv"

df = pd.read_csv(csv_filepath)
```

```python
df
```

### How to display starting salary for each degree

```python
df_ss = df.iloc[:, 0:2] # Data for major and starting salary 
```

Convert salary from string with '$' to float number

```python
df_ss.iloc[:, 1] = df_ss.iloc[:, 1].replace('[\$,]', '', regex=True).astype(float)

df_ss
```

# How to plot a bar graph using pandas

Reference: 
- https://www.youtube.com/watch?v=AFnUhiRXXWk

```python
%matplotlib inline  

df_ss.plot.bar(x="Undergraduate Major")
```

### How to customise the plot

```python
%matplotlib inline  
plt.rcParams['figure.figsize'] = (20,14)

ax = df_ss.plot.bar(
    x="Undergraduate Major",
    y="Starting Median Salary",
    title="V comp",
    color=list('rgbkym'),
    legend=True
)

ax.set_xlabel("Degree", fontsize=16)
ax.set_ylabel("Salary ($)", fontsize=12)
```
